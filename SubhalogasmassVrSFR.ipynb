```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
h = 0.6711

def SubhalogasmassVrSFR(filePathSnap, filePathfof, color):
    with h5py.File(filePathSnap, "r") as data, \
         h5py.File(filePathfof, "r") as f:
        
        total_mass = f["Subhalo/SubhaloMass"][:] * 1e10 / h  # Msun
        sfr = f["Subhalo/SubhaloSFR"][:]  # Msun/yr

        plt.scatter(total_mass, sfr, s=4, c=color)
        plt.xscale("log")
        plt.yscale("log")
        plt.xlabel("Subhalo Mass (Msun)")
        plt.ylabel("Star Formation Rate (Msun/yr)")
        plt.title("Subhalo Mass vs Star Formation Rate")
        plt.tight_layout()
```

```python
snap_simba_z3 = "data/Output-simba/snap_004.hdf5"
fof_simba_z3  = "data/Output-simba/fof_subhalo_tab_004.hdf5"
snap_tng_z3   = "data/Output-fidTNG/snap_004.hdf5"
fof_tng_z3    = "data/Output-fidTNG/fof_subhalo_tab_004.hdf5"

snap_simba_z0 = "data/Output-simba/snap_032.hdf5" 
fof_simba_z0  = "data/Output-simba/snap_032.hdf5"
snap_tng_z0   = "data/Output-fidTNG/fof_subhalo_tab_032.hdf5"
fof_tng_z0    = "data/Output-fidTNG/fof_subhalo_tab_032.hdf5"

plt.figure(figsize=(6,4))
SubhalogasmassVrSFR(snap_simba_z3, fof_simba_z3, "red")
plt.title("BH Mass Distribution (z â‰ˆ 3)")
plt.show()
```
